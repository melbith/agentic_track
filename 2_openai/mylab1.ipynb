```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
load_dotenv(override=True)
```

```python
# Create an agent with Name, Instructions, and a Model
agent = Agent(
    name="Jokester",
    instructions="You are a joke teller. Your job is to tell jokes.",
    model="gpt-4o-mini"
)

print(agent)
```

```python
# The wrong way to run the agent.
result = Runner.run(agent, "Tell a joke about Autonomous AI Agents?")
result
```

```python
# The right way - Run the Agnet with Runner.run(agent, user prompt) then print final_output
# This would have the agent interactions recorded with the General heading "Agent workflow"
result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents?")
print(result.final_output)
```

Check the traces @ https://platform.openai.com/traces

```python
# Now wrap these in a trace so that the agent interactions get recorded with the heading "Telling a joke"
with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents?")
    print(result.final_output)
```
