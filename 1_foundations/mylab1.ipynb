```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
import os
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    print(f"API key loaded and begins with: {openai_api_key[:8]}...")
else:
    print("OpenAI API key not set")
```

```python
# 1. Import the OpenAI class
from openai import OpenAI

# 2. Create a client object to interact with the API
#    (It automatically looks for your API key in an environment variable
#    called OPENAI_API_KEY)
openai_client = OpenAI()

# 3. Create a list of messages in the familiar OpenAI format
messages=[
{"role": "user", "content": "What is 5*8?"}
]

# 4. Use the client to make a request, for example, to the chat model
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

# 5. Print the model's answer
print(response.choices[0].message.content)
```

```python
message = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages=[
{"role": "user", "content": message}
]
```

```python
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

question = response.choices[0].message.content
print(question)
```

```python
messages=[
{"role": "user", "content": question}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

answer = response.choices[0].message.content
print(answer)
```

```python
from IPython.display import Markdown, display
display(Markdown(answer))
```

            First ask the LLM to pick a business area that might be worth exploring for an Agentic AI opportunity.<br/>
            Then ask the LLM to present a pain-point in that industry - something challenging that might be ripe for an Agentic solution.<br/>

```python
# 1. Indentify a business area that might be worth exploring for an Agentic AI opportunity.

message = "Pick a business area that might be worth exploring for an Agentic AI opportunity. Respond only with the business area."
messages = [
{"role": "user", "content": message}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

business_area = response.choices[0].message.content
print(business_area)
```

```python
# 2. Present a pain-point in that industry - something challenging that might be ripe for an Agentic solution.

message = "Present a pain-point in that industry - something challenging that might be ripe for an Agentic solution."
messages = [
{"role": "user", "content": message}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

pain_point = response.choices[0].message.content
print(pain_point)
```

```python
# 3. Propose the Agentic AI solution for the pain-point.

message = "Propose the Agentic AI solution for this pain-point."
messages = [
{"role": "user", "content": message}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

solution = response.choices[0].message.content
print(solution)
```
